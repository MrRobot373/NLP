```python
document = [(movie_reviews.words(file_id),category) for file_id in movie_reviews.fileids() for category in movie_reviews.categories(file_id)]
document
```

```python
def find_feature(word_list, feature_vector):
    features = {}

    for x in feature_vector:
        features[x] = x in word_list

    return features
```

```python
import numpy as np

def find_feature(word_list, feature_vector):
    # Assuming feature_vector is a NumPy array
    return np.isin(feature_vector, word_list)

# Convert feature_vector to a NumPy array if it's not already
feature_vector_np = np.array(feature_vector)

# Vectorized approach using NumPy
feature_sets = [(find_feature(word_list, feature_vector_np), category) for (word_list, category) in document]
```

SVM

```python
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.svm import SVC
from sklearn import model_selection

train_set,test_set = model_selection.train_test_split(feature_sets,test_size = 0.25)

print(len(train_set))
print(len(test_set))
```

```python
from nltk.classify import SklearnClassifier
from sklearn.svm import SVC
from sklearn.feature_extraction import DictVectorizer
import nltk

# Sample data
data_point_1 = ({'feature1': 0.2, 'feature2': 0.8}, 'positive')
data_point_2 = ({'feature1': 0.7, 'feature2': 0.3}, 'negative')

# Example train_set and test_set
train_set = [data_point_1, data_point_2]
test_set = [data_point_1, data_point_2]  # In a real scenario, this would be a separate set

# Train the model
model = SklearnClassifier(SVC(kernel='linear'))
model.train(train_set)

# Evaluate accuracy
accuracy = nltk.classify.accuracy(model, test_set)
print('SVC Accuracy: {}'.format(accuracy))
```
